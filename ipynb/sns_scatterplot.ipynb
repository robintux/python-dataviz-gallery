```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
np.random.seed(0)
```

```python
n = 1000
d = {'x': np.random.randn(n),
     'y': np.random.randn(n)}
```

```python
df = pd.DataFrame(d)
```

```python
fig = sns.jointplot(x='x', y='y', data=df)
```

```python
fig.savefig('../images/sns_scatterplot.png', dpi=80)
```
